```python
'''
performing template detection for the given image and detecting given image in the given template
'''
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load main image and template
img = cv2.imread(r"D:\kathan\python_codes\7th sem\CV lab\test_img.jpg")        # image with multiple persons
template = cv2.imread(r"D:\kathan\python_codes\7th sem\CV lab\image.png")      # image of the person

# Convert to grayscale
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray_template = cv2.cvtColor(template, cv2.COLOR_BGR2GRAY)

# Get template size
w, h = gray_template.shape[::-1]

# Perform template matching
result = cv2.matchTemplate(gray_img, gray_template, cv2.TM_CCOEFF_NORMED)

# Threshold to filter good matches
threshold = 0.8
locations = np.where(result >= threshold)   #  Use NumPy instead of cv2

# Draw rectangles for all detected matches
for pt in zip(*locations[::-1]):  # Switch x and y
    cv2.rectangle(img, pt, (pt[0] + w, pt[1] + h), (0, 255, 0), 2)

# Show result
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Template Detection")
plt.axis("off")
plt.show()
```
