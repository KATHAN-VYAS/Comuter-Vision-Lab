```python
'''argument inside thumbnail is maximum size of picture'''

from PIL import Image
import matplotlib.pyplot as plt

# Open the original image
img2 = Image.open('image.png')

# Make a copy and resize it
cp_img2 = img2.copy()
cp_img2.thumbnail((200, 200))  # Resize in place to max 200x200

# Display with matplotlib
plt.imshow(cp_img2)
plt.axis('off')
plt.title('Thumbnail Image')
plt.show()
```

```python
Gray_img = img2.convert("L")


r,g,b = img2.split()
plt.Figure(figsize=(25,10))
plt.subplot(1,4,1)
plt.imshow(r, cmap="Reds")
plt.axis("off")
plt.subplot(1,4,2)
plt.imshow(g, cmap="Greens")
plt.axis("off")
plt.subplot(1,4,3)
plt.imshow(b, cmap="Blues")
plt.axis("off")
plt.subplot(1,4,4)
plt.imshow(r)
plt.axis("off")
plt.show()
```

```python
### when we convert a channel of an image to grayscale using cmap="gray" its not will be in grayscale
# but with incorrect pixel values


plt.figure(figsize = (15,8))
plt.subplot(1,2,1)
plt.imshow(Gray_img, cmap = "gray")
plt.subplot(1,2,2)
plt.imshow(r, cmap = "gray")
plt.show()
```
