```python
import cv2
import matplotlib.pyplot as plt

img1 = cv2.imread(r'D:\kathan\python_codes\7th sem\CV lab\Comuter-Vision-Lab\image.png')
plt.imshow(img1)

image_type = type(img1)
shape = img1.shape
img_height = shape[0]
img_width = shape[1]
channels= shape[2]
data_type = img1.dtype
size = img1.size
min_pixel_val = img1.min()
max_pixel_val = img1.max()
print("image type:",image_type)
print("imgage shape:",shape)
print('Image height:',img_height)
print('image width', img_width)
print("no of channels",channels )
print("data type", data_type)
print("image size", size)
print("minimum pixel value", min_pixel_val)
print("maximum pixel value", max_pixel_val)


#split into channel

b,g,r = cv2.split(img1)
print("blue channel shape", b.shape)

print("Green channel shape:", g.shape)
print("Red channel shape:", r.shape)

# Average pixel value per channel
mean_bgr = cv2.mean(img1)
print("\n=== Mean Pixel Values ===")
print(f"Mean Blue: {mean_bgr[0]:.2f}")
print(f"Mean Green: {mean_bgr[1]:.2f}")
print(f"Mean Red: {mean_bgr[2]:.2f}")

gray = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(gray)
print("\n=== Min/Max in Grayscale ===")
print(f"Min value: {min_val} at location {min_loc}")
print(f"Max value: {max_val} at location {max_loc}")
```

```python
#converting image to new formate

img_rgb= cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
cv2.imwrite("converted image.jpg",img1 )
```

```python
# croping image:
cropped_img = img_rgb[20:150, 100:400]
plt.imshow(cropped_img)
```
